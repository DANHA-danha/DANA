# NICE평가정보 2024년 상반기 신용평가 데이터 분석

```python
import pandas as pd

df = pd.read_csv('credit_data_2024H1.csv')
print(f'데이터 로드 완료: 총 {len(df):,}건')
print(f'분석 기간: {df["date"].min()} ~ {df["date"].max()}')
```

## 1. 신용등급 분포 현황

```python
grade_dist = df['credit_grade'].value_counts().sort_index()
print('신용등급별 분포:')
for grade, count in grade_dist.items():
    print(f'{grade}: {count:,}건 ({count/len(df)*100:.1f}%)')
```

## 2. 업종별 신용위험 분석

```python
sector_pd = df.groupby('sector')['default_prob'].mean()
print('업종별 평균 부도율 (PD):')
for sector, pd_rate in sector_pd.items():
    print(f'{sector}: {pd_rate:.2f}%')
```

## 3. 월별 신규 평가 건수 추이

```python
monthly = df.groupby(df['date'].str[:7]).size()
print('월별 신규 평가 건수:')
for month, count in monthly.items():
    print(f'{month}: {count:,}건')
```

## 4. 등급 변동(Migration) 분석

```python
migration = df[df['is_reassessment']==True]
upgrade = migration[migration['grade_change']>0]
downgrade = migration[migration['grade_change']<0]
print(f'등급 변동 현황 (재평가 대상 {len(migration):,}건):')
print(f'- 등급 상향: {len(upgrade):,}건')
print(f'- 등급 하향: {len(downgrade):,}건')
```

## 5. 핵심 요약
- 전체 신용등급 분포는 전년과 유사하나, 하위등급 비율 소폭 증가 추세
- 부동산업(+0.91%p)과 음식숙박업(+0.78%p) 부도율 상승이 두드러짐
- 신규 평가 건수는 전년 대비 4.9% 증가, 3월이 최다
- 등급 하향 비율이 상향을 소폭 초과하여 전반적 신용위험 상승 시그널
